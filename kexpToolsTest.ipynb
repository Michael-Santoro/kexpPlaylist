Being a big fan of KEXP and especially the morning show I am going to use the api found here to scrape song information from the show. https://api.kexp.org/v2/

```python
import requests
import pandas as pd
from kexpTools import kexp
```

```python
kexp = kexp()
```

```python
todaysShows = kexp.getShowsDay(2022,8,10)
```

```python
morningShow = todaysShows[todaysShows.program_name=='The Morning Show']
```

```python
plays_7am = kexp.getSongsHour(2022,8,10,7,7)
```

```python
plays_7am
```

```python
playlist = plays_7am[['song','artist', 'album']]
```

```python
playlist.rename(columns={'song':'Name','artist':'Artist','album':'Album'},inplace=True)
```
