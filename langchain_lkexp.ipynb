```python
import requests
import pandas as pd

# Define the base URL of the API
base_url = 'https://api.kexp.org/v2/timeslots/?format=json'

# Make a GET request to the resource URL to retrieve a list of items
response = requests.get(base_url)

# Check the response status code
if response.status_code == 200:
    # If the response is successful, print the list of items
    items = response.json()
else:
    # If the response is unsuccessful, print the error message
    print(f"Error: {response.status_code} - {response.reason}")
```

# Prompt JSON Time Query

```python
def format_time(date_str):
        date_time_obj = datetime.strptime(date_str, '%Y-%m-%d %H:%M:%S')
        return date_time_obj.isoformat() + 'Z'

def make_request(loc):
        # Make a GET request to the resource URL to retrieve a list of items
        response = requests.get(loc)

        # Check the response status code
        if response.status_code == 200:
                # If the response is successful, print the list of items
                return response.json()
                
        else:
                # If the response is unsuccessful, print the error message
                print(f"Error: {response.status_code} - {response.reason}")
```

```python
date_start = '2022-04-13 07:00:00'
date_stop = '2022-04-14 07:00:00'

root_str = 'https://api.kexp.org/v2/plays/?airdate_after=<<start_date>>&airdate_before=<<end_date>>&exclude_airbreaks=true'
```

```python
result = make_request(root_str.replace('<<start_date>>',format_time(date_start)).replace('<<end_date>>',format_time(date_stop)))
```

```python
result
```

```python
json_spec = JsonSpec(dict_=result, max_value_length=4000)
json_toolkit = JsonToolkit(spec=json_spec)

json_agent_executor = create_json_agent(
    llm=OpenAI(temperature=0),
    toolkit=json_toolkit,
    verbose=True
)
```

```python
_result = json_agent_executor.run("List the song artist and album that were played?")
```
